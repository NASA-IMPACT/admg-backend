```python
import pandas as pd
from ingest import sheet
```

```python
# this should be a download of the inventory sheet
gmcd_file_path = 'gcmd_convert_raw.xlsx'
gcmd_raw_file = pd.read_excel(gmcd_file_path, sheet_name = None)
```

```python
# this should be a text file with a copy/pasted column from the inventory sheet
column_file_path = 'gcmd_column.txt'
with open(column_file_path, 'r') as f:
    column = [line.rstrip() for line in f] 
```

```python
mapping_columns = {
    "gcmd": {
        "Internal 4-digit code": "ignore_code",
        "Category": "category",
        "Topic": "topic",
        "Term": "term",
        "Variable Level 1": "variable_1",
        "Variable Level 2": "variable_2",
        "Variable Level 3": "variable_3",
        "UUID": "gcmd_uuid"
  },
}
```

```python
gcmd_new = sheet(
    excel_data=gcmd_raw_file, 
    remap_dict=mapping_columns,
    sheet_name='GCMD Phenomena - No Services',
    remap_name='gcmd',
    header_row=5, 
    data_row=6
)
gcmd_old = sheet(
    excel_data=gcmd_raw_file, 
    remap_dict=mapping_columns,
    sheet_name='GCMD Phenomena',
    remap_name='gcmd',
    header_row=5, 
    data_row=6
)
```

```python
mapping_old = {'':''}
for index, row in gcmd_old.iterrows():
    code = str(row['ignore_code'])
    uuid = row['gcmd_uuid']
    mapping_old[code]=uuid
    
mapping_new = {'':''}
for index, row in gcmd_new.iterrows():
    code = str(row['ignore_code'])
    uuid = row['gcmd_uuid']
    mapping_new[uuid]=code
```

```python
new_column = []
for line in column:
    line_parsed = [item.strip().replace('"','') for item in line.split(',')]
    line_uuids= [mapping_old[old_code] for old_code in line_parsed]
    line_new_codes = [mapping_new[old_uuid] for old_uuid in line_uuids]
    line_new_text = ', '.join(line_new_codes)
    new_column.append(line_new_text)
```

```python
with open('column_new.txt', 'w') as f:
    f.writelines("%s\n" % place for place in new_column)
```
